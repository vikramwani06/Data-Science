## Normal Distribution

<li>The normal distribution is a continuous distribution or a function that can take on values anywhere on the real
line.
<li>The normal distribution takes two parameters- the mean of the distribution μ and the variance(sigma) σ2.

```python
from scipy.stats import norm
import matplotlib.pyplot as plt
import numpy as np
```

```python
random = norm.rvs(loc=0,scale=1,size=400)
parameters = norm.fit(random)
x = np.linspace(-5,5,100)
fitted_distribution = norm.pdf(x,loc = parameters[0],scale = parameters[1])
normal_distribution = norm.pdf(x)
```

```python
plt.plot(x,fitted_distribution,"red",label="Fitted normal dist",linestyle="dashed", linewidth=2)
plt.plot(x,normal_distribution,"blue",label="Normal distribution",linewidth=2)
plt.hist(random,normed=1,color="cyan",alpha=.3) #alpha, from 0 (transparent) to 1 (opaque)
plt.title("Normal distribution fitting")
plt.legend()
plt.show()
```

<li>The above block of code generates an array of 400 random samples from a normal distribution with mean 0
and variance 1
<li>The norm class from scipy library creates the random variables using norm.rvsnorm.fit(data) returns a list of two parameters (mean, parameters[0] and std, parameters[1]), which should be
in array form.
<li>The two parameters are mean and the variance of the fitted distribution
<li>Now we generate the normal and fitted distributions and plot them

```python
import numpy as np
# Sample from a normal distribution using numpy's random number generator
samples = np.random.normal(size=10000)
```

```python
# Compute a histogram of the sample
bins = np.linspace(-5, 5, 30)
histogram, bins = np.histogram(samples, bins=bins, normed=True)
bin_centers = 0.5*(bins[1:] + bins[:-1])
```

```python
# Compute the PDF on the bin centers from scipy distribution object
from scipy import stats
pdf = stats.norm.pdf(bin_centers)
```

```python
from matplotlib import pyplot as plt
plt.figure(figsize=(6, 4))
plt.plot(bin_centers, histogram, label="Histogram of samples")
plt.plot(bin_centers, pdf, label="PDF")
plt.legend()
plt.show()
```

<li>The above block of code represents a histogram built from samples and the PDF (probability density function).
<li>Here, we create the random samples using numpy's random number generator
<li>The stats class from the scipy library computes the probability density function bin centers
<li>The matplotlib plots the histogram of the samples

```python
mu, sigma = 0, 0.1 # mean and standard deviation
s = np.random.normal(mu, sigma, 1000)
```

```python
abs(mu - np.mean(s)) < 0.01
```

```python
abs(sigma - np.std(s, ddof=1)) < 0.01
```

```python
import matplotlib.pyplot as plt
count, bins, ignored = plt.hist(s, 30, normed=True)
plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) *
 np.exp( - (bins - mu)**2 / (2 * sigma**2) ),
 linewidth=2, color='r')
plt.show()
```

<li>Normal distribution can take values from minus infinity to plus infinity.

```python
from scipy.stats import norm
import matplotlib.pyplot as plt
fig, ax = plt.subplots(1, 1)
```

```python
mean, var, skew, kurt = norm.stats(moments='mvsk')
```

```python
x = np.linspace(norm.ppf(0.01),
 norm.ppf(0.99), 100)
ax.plot(x, norm.pdf(x),
 'r-', lw=5, alpha=0.6, label='Normal Density')
```

<li>The above block of code displays the probability density function

```python
rv = norm()
ax.plot(x, rv.pdf(x), 'k-', lw=2, label='Frozen Density')
```

```python
vals = norm.ppf([0.001, 0.5, 0.999])
np.allclose([0.001, 0.5, 0.999], norm.cdf(vals))
```

<li>The above block of code checks the accuracy of the samples in the normal distribution

```python
r = norm.rvs(size=500)
```

```python
ax.hist(r, normed=True, histtype='stepfilled', alpha=0.2)
ax.legend(loc='best', frameon=False)
plt.show()
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(x, norm.cdf(x))
```

```python
accuracy= 100-mean_squared_error(x,norm.cdf(x))
accuracy
```

<li>The above block of code computes the mean squared error of the distribution based on which we calculate the accuracy which turns out to be 98%
<li>Hence, we can conclude that the distribution is accurate

## Exponential Distribution

<li>As an instance of the rv_continuous class, expon object inherits from it a collection of generic methods and completes them with details specific for this particular distribution.
<li>The probability density above is defined in the “standardized” form. To shift and/or scale the distribution use the loc and scale parameters. Specifically, expon.pdf(x, loc, scale) is identically equivalent to expon.pdf(y) / scale with y = (x - loc) / scale.
<li>A common parameterization for expon is in terms of the rate parameter lambda, such that pdf = lambda exp(-
lambda x). This parameterization corresponds to using scale = 1 / lambda.


```python
from scipy.stats import expon
import matplotlib.pyplot as plt
fig, ax = plt.subplots(1, 1)
```

```python
mean, var, skew, kurt = expon.stats(moments='mvsk')
```

```python
x = np.linspace(expon.ppf(0.01),
 expon.ppf(0.99), 100)
ax.plot(x, expon.pdf(x),
 'r-', lw=5, alpha=0.6, label='Exponential Density')
```

```python
rv = expon()
ax.plot(x, rv.pdf(x), 'k-', lw=2, label='Frozen Density')
```

```python
vals = expon.ppf([0.001, 0.5, 0.999])
np.allclose([0.001, 0.5, 0.999], expon.cdf(vals))
```

<li>The above block of code checks the accuracy of the Cumulative Density Function and the Percent Point Function

```python
r = expon.rvs(size=1000)
```

```python
ax.hist(r, normed=True, histtype='stepfilled', alpha=0.2)
ax.legend(loc='best', frameon=False)
plt.show()
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(x, expon.cdf(x))
```

```python
accuracy= 100-mean_squared_error(x,expon.cdf(x))
accuracy
```

<li>Here, we predict the accuracy of the distribution, which turns out to be 96%
<li>Hence we can conclude that the distribution is accurate

## Bernoulli Distribution

<li>Discrete random variables are defined from a standard form and may require some shape parameters to complete its specification.
<li>Any optional keyword parameters can be passed to the methods of the RV object

```python
from scipy.stats import bernoulli
import matplotlib.pyplot as plt
fig, ax = plt.subplots(1, 1)
```

```python
p = 0.3
mean, var, skew, kurt = bernoulli.stats(p, moments='mvsk')
```

```python
x = np.arange(bernoulli.ppf(0.01, p),
 bernoulli.ppf(0.99, p))
ax.plot(x, bernoulli.pmf(x, p), 'bo', ms=8, label='Bernoulli Density')
ax.vlines(x, 0, bernoulli.pmf(x, p), colors='b', lw=5, alpha=0.5)
```

```python
rv = bernoulli(p)
ax.vlines(x, 0, rv.pmf(x), colors='k', linestyles='-', lw=1,
 label='Frozen Density')
ax.legend(loc='best', frameon=False)
plt.show()
```

```python
prob = bernoulli.cdf(x, p)
np.allclose(x, bernoulli.ppf(prob, p))
```

<li>The above block of code checks the accuracy of the Cumulative Density Function and the Percent Point Function

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(x, bernoulli.cdf(x,p))
```

```python
accuracy= 100-mean_squared_error(x,bernoulli.cdf(x,p))
accuracy
```

<li>As the mean squared error in this case is 0.48 and also the accuracy is 99%, we can conclude that the distribution is accurate
